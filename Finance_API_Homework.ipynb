Task 2.1: Fetch Additional Data Using Modules

```python
import json
import pandas as pd
import json
import requests
url = "https://yfapi.net/v6/finance/quote"

stock = input("Enter stock symbol (e.g., AAPL, MSFT): ")

apikey="Rmp6k8gV1w8fAmFQBdwub9EEy6b2GV0T1g03qskX"

querystring = {
    "symbols": stock
}
headers = {
    'x-api-key': apikey
    }

response = requests.request("GET", url, headers=headers,params=querystring)

data = response.json()

if "quoteResponse" in data and "result" in data["quoteResponse"] and len(data["quoteResponse"]["result"]) > 0:
  stock_data = data["quoteResponse"]["result"][0]
else:
    print("Error: Invalid stock symbol or module. Please check the API response.")
    exit()

print("Choose a module:")
print("1)52 Week high")
print("2)52 Week Low")
print("3)Return on Assets (ROA)")

choice = input("Enter your choice (1/2/3): ")

if choice == "1":
    module = "fiftyTwoWeekHigh"
    stock_data.get("fiftyTwoWeekHigh", "N/A")
elif choice == "2":
    module = "fiftyTwoWeekLow"
    stock_data.get("fiftyTwoWeekLow", "N/A")
elif choice == "3":
    module = "returnOnAssets"
    stock_data.get("returnOnAssets", "N/A")
else:
    print("Invalid choice. Please enter 1, 2, or 3.")
    exit()

df = pd.DataFrame({
    "Metric": [module],
    "Value": [stock_data.get(module, "N/A")]
})

print(df)
```
